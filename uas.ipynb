```python
import cv2
import matplotlib.pyplot as plt

def crop_image(image, x, y, width, height):
    cropped_image = image[y:y+height, x:x+width]
    return cropped_image

def rotate_image(image, angle):
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated_image = cv2.warpAffine(image, M, (w, h))
    return rotated_image

# Load image
image_path = 'dinda.jpg'  # Ubah dengan path gambar yang ingin Anda gunakan
image = cv2.imread(image_path)

# Tampilkan gambar asli
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('dinda.jpg')
plt.axis('off')
plt.show()

# Crop image
x, y, width, height = 100, 100, 300, 200  # Ganti nilai sesuai kebutuhan Anda
cropped_image = crop_image(image, x, y, width, height)

# Tampilkan gambar yang dipotong
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')
plt.axis('off')
plt.show()

# Rotate image
angle = 45  # Sudut rotasi (dalam derajat), ganti sesuai kebutuhan Anda
rotated_image = rotate_image(image, angle)

# Tampilkan gambar yang diputar
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')
plt.axis('off')
plt.show()
```
